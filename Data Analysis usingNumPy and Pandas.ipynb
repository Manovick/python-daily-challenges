```python
import numpy as np
import pandas as pd
```

NumPy Series

```python
temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])
print("Week 1 Temperature:",temperatures_w1)
```

```python
print("shape:",temperatures_w1.shape)
print("Data type:",temperatures_w1.dtype)
print("Number of elements:",temperatures_w1.size)
```

```python
temperature_f = (temperatures_w1 *9/5)+32
print("Temperatures in Fharenheit:",temperature_f)
```

```python
max_temp = np.max(temperatures_w1)
min_temp = np.min(temperatures_w1)
mean_temp = np.mean(temperatures_w1)

print("maximum Temperature:",max_temp)
print("minimum Temperature:",min_temp)
print("Average Temperature: {:.2f}".format(mean_temp))
```

```python
first_three_days = temperatures_w1[:3]
print("First three days:",first_three_days)
```

```python
weekend = temperatures_w1[-2:]
print("Weekend Temperature:",weekend)
```

```python
middle_three_days = temperatures_w1[2:5]
print('Middle three days:',middle_three_days)
```

```python
temperatures = np.array([
    [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9],
    [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]
])

print("2D Temperature Array:",temperatures)
```

```python
print("Shape:",temperatures.shape)
print("Data type;",temperatures.dtype)
print("Number of elements:",temperatures.size)
```

```python
week1 = temperatures[0]
print("Week 1:",week1)
```

```python
week2 = temperatures[1]
print("Week 2:",week2)
```

```python
weekend_both_weeks = temperatures[:, -2:]
print("Weekend Temperatures for (Both Weekends):",weekend_both_weeks)
```

Pandas Series

```python
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

marks
```

```python
first_rank_mark = marks.iloc[0]
print("Mark of Rank1:",first_rank_mark)
```

```python
top_3 = marks.loc[['Rank1','Rank2','Rank3']]
print("Top 3 ranks:\n",top_3)
```

```python
third_rank = marks.iloc[2]
print("Third rank student mark:",third_rank)
```

```python
high_mark = marks[marks > 90]
print("students high marks:\n",high_mark)
```

```python
marks.loc['Rank1'] =100
print("Modify mark:\n",marks)
```

```python
marks = marks.drop('Rank5')
```

```python
cgpa = marks / 10
print("CGPA:\n",cgpa)
```
